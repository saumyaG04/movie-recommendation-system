```python
import seaborn as snspip
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load cleaned data
data = pd.read_csv(r"C:\Users\saumy\OneDrive\Documents\Python Projects\portfolio\movie recommendation system\notebooks\data\cleaned_data.csv")
```

```python
# Visualize popular movies
popular_movies = data.groupby("title").size().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 5))
sns.barplot(y=popular_movies.index, x=popular_movies.values, palette="viridis")
plt.title("Top 10 Popular Movies")
plt.xlabel("Number of Ratings")
plt.show()
```

```python
# Visualize average ratings
avg_ratings = data.groupby("title")["rating"].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 5))
sns.barplot(y=avg_ratings.index, x=avg_ratings.values, palette="coolwarm")
plt.title("Top 10 Highest Rated Movies")
plt.xlabel("Average Rating")
plt.show()
```
